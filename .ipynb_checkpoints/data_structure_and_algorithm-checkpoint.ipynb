Title case algorithm

```python
def title_case(s: str):
    words = s.lower().split(" ")
    for i in range(len(words)):
        words[i] = words[i][0].upper() + words[i][1:]
    return " ".join(words)

print(title_case("hello world from south sudan"))
```
